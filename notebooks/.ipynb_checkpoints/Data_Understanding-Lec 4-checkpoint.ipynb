```python
page=requests.get('https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Fallzahlen.html')
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
html_table=soup.find('table')
```

```python
all_rows=html_table.find_all('tr')
```

```python
final_data=[]
```

```python
for pos,rows in enumerate(all_rows):
    #print(pos)
    #print(rows)
    col_list=[each_col.get_text(strip=True) for each_col in rows.find_all('td')]
    final_data.append(col_list)
    
```

```python

pd.DataFrame(final_data)
```

```python
page1=requests.get('https://www.worldometers.info/coronavirus/#countries')
```

```python
soup = BeautifulSoup(page1.content,'html.parser')
```

```python
html_tablewrld=soup.find('table')
```

```python
all_rowswrld = html_tablewrld.find_all('tr')
```

```python
data = []
for pos, rows in enumerate(all_rowswrld):
    #print(rows)
    col_lst = [each_col.get_text(strip=True) for each_col in rows.find_all('td')]
    data.append(col_lst)
```

```python
df=pd.DataFrame(data).dropna().rename(columns={0:'countries'})
df.drop([1,2,3,4,5,6,7,8])
```

```python
pageIndia = requests.get('https://www.mohfw.gov.in/')
soupI= BeautifulSoup(pageIndia.content,'html.parser')
html_india = soupI.find('table')
html_row =soupI.find_all('tr')
data_I=[]
for pos, row in enumerate(html_row):
    col_Ind = [each_col.get_text(strip=True) for each_col in row.find_all('td')]
    data_I.append(col_Ind)
    
```

```python
dff = pd.DataFrame(data_I).dropna().rename(columns={1:'States',2:'Total Confirmed cases',3:'Cured',4:'Death'})
dff.to_csv('file_name.csv')
```
