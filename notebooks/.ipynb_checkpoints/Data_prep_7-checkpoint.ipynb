```python
pd_JH_data= pd.read_csv('..\data\processed\COVID_relational_confirmed.csv',sep= ';', parse_dates=[0])
pd_JH_data=pd_JH_data.sort_values('date',ascending=True).reset_index(drop=True).copy()
pd_JH_data.head()
```

```python
test_data= pd_JH_data[((pd_JH_data['country']=='Canada')|
                      (pd_JH_data['country']=='India'))&
                     (pd_JH_data['date']>'2020-04-20')]
```

```python
import numpy as np
from sklearn import linear_model
reg = linear_model.LinearRegression(fit_intercept=True)
import pandas as pd

from scipy import signal


def get_doubling_time_via_regression(in_array):
    ''' Use a linear regression to approximate the doubling rate
        Parameters:
        ----------
        in_array : pandas.series
        Returns:
        ----------
        Doubling rate: double
    '''

    
    #y = np.array(in_array['confirmed'])
    #X = np.arange(len(y)).reshape(-1, 1)
    y = np.array(in_array)
    X = np.arange(-1,2).reshape(-1, 1)

    #assert len(in_array)==3
    reg.fit(X,y)
    intercept=reg.intercept_
    slope=reg.coef_

    return intercept/slope




if __name__ == '__main__':
    test_data_reg=np.array([2,4,6])
    result=get_doubling_time_via_regression(test_data_reg)
    print('the test slope is: '+str(result))
```

```python
test_data.groupby(['state','country']).agg(np.max)
```

```python
test_data.groupby(['state','country']).apply(get_doubling_time_via_regression)
```

```python
def rolling_reg(df_input,col='confirmed'):
    
    days_back=3
    result=df_input['confirmed'].rolling(window=days_back,min_periods=days_back).apply(get_doubling_time_via_regression,raw=False)
    return result
```

```python
test_data[['state','country','confirmed']].groupby(['state','country']).apply(rolling_reg,'confirmed')
```
